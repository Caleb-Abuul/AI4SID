# Currency Converter

This program converts five of the most popular currencies (US Dollar (USD), Canadian Dollar (CAD), Euro, Pound Sterling (GBP), and Japanese Yen(JPY)) in the world to the Nigerian Naira (NGN), based on the current exchange rate as of February 17, 2026.

Yes I know! Currencies fluctuate on a daily basis.

However, this is a mini-project, so no APIs are allowed. So, I will do my best to keep it as up-to-date as possible. 

I will update the currency rates weekly ðŸ“†

```python
# accepts two inputs from the user, the currency and the amount and stores them in the variables 'currency' and 'amount' respectively. 
currency = int(input("Chose currency: \n1. US Dollar \n2. Canadian Dollar \n3. Euro \n4. Pound Sterling \n5. Japanese Yen"))
amount = int(input("Enter the amount"))

# initializes two variables, 'rate' and 'result', to 0. 
rate = 0
result = 0

def convertToNaira(cur, amt):
    # accepts two parameters, the chosen currency and amount, and uses a series of if-elif statements to determine the exchange rate based on the chosen currency. 

    if cur == 1:
        rate = 1354.86
        result = amt * rate
        print(f"{amt} USD equals {result} NGN, at {rate} NGN per USD")
    elif cur == 2:
        rate = 992.98
        result = amt * rate
        print(f"{amt} CAD equals {result} NGN, at {rate} NGN per CAD")
    elif cur == 3:
        rate = 1605.20
        result = amt * rate
        print(f"{amt} Euro equals {result} NGN, at {rate} NGN per Euro")
    elif cur == 4:
        rate = 1844.22
        result = amt * rate
        print(f"{amt} GBP equals {result} NGN, at {rate} NGN per GBP")
    elif cur == 5:
        rate = 8.85
        result = amt * rate
        print(f"{amt} JPY equals {result} NGN, at {rate} NGN per JPY")
    else:
        print("invalid input")
        return 0
    
# finally, calls the function
convertToNaira(currency, amount)
```
